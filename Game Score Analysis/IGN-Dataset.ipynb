```python
Index1=['Above Average','Below Average','Total Games']
Index=['Above Average','Below Average']
platform=['PC','Xbox 360','PlayStation 3','PlayStation 4','Xbox One']
avgscore=data['score'].mean()

print(avgscore)

result = np.zeros((3,len(platform)))
percentage=np.zeros((2,len(platform)))
for i in range(len(platform)):
    aboveavg = (data['platform']==platform[i]) & (data['score']>=avgscore)
    result[0,i]=data[aboveavg].shape[0]
    belowavg = (data['platform']==platform[i]) & (data['score']<=avgscore)
    result[1,i]=data[belowavg].shape[0]
    result[2,i]=result[0,i]+result[1,i]
    percentage[0,i]=(result[0,i]/result[2,i])*100
    percentage[1,i]=(result[1,i]/result[2,i])*100
#print(result)
#print(percentage)

resultFrame1 = pd.DataFrame(result,index=Index1,columns=platform)
print(resultFrame1)
resultFrame = pd.DataFrame(percentage,index=Index,columns=platform)
print(resultFrame)
resultFrame.plot(kind='bar')
newDataFrame=resultFrame.T
newDataFrame.plot(kind='bar')
```

```python
# global mean 

mn = data['score'].mean()
print(mn)

plat=['PC','Xbox 360','PlayStation 3','PlayStation 4','Xbox One']

result = np.zeros((3,len(plat)))
per = np.zeros((2,len(plat)))

for i in range(len(plat)):
    result[0,i] = data[(data['platform']==plat[i]) & (data['score']>=mn)].shape[0] # above average
    result[1,i] = data[(data['platform']==plat[i]) & (data['score']<mn)].shape[0] # below average
    result[2,i] = result[0,i]+result[1,i]
    per[0,i] = (result[0,i]/result[2,i])*100
    per[1,i] = (result[1,i]/result[2,i])*100
    

print(result) 
print(per)
```
