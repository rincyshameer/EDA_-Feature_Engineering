# EDA and Feature Engineering on Flight_Price_prediction Dataset

```python
#Importing the libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Importing the train dataset 
df_train=pd.read_excel('Data_Train.xlsx')
df_train.head()
```

```python
df_test=pd.read_excel('Test_set.xlsx')
df_test.head()
```

```python
#Combining the train and test dataset

df=df_train.append(df_test)
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
#Feature Engineering 

#Inorder to convert the Date of journey into day,month,year sepaeratly- 1sr method

df['Date']=df['Date_of_Journey'].str.split('/').str[0] #to extract each value from the string
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]


#2nd method- using lamba functions

df['Date']=df['Date_of_Journey'].apply(lambda x:x.split('/')[0])
df['Month']=df['Date_of_Journey'].apply(lambda x:x.split('/')[1])
df['Year']=df['Date_of_Journey'].apply(lambda x:x.split('/')[2])

```

```python
df.head()
```

```python
#converting the date,month,year string formats into integer value

df['Date']=df['Date'].astype('int64')
df['Month']=df['Month'].astype('int64')
df['Year']=df['Year'].astype('int64')
```

```python
df.info()
```

```python
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df.head()
```

```python
#Handling Arrival time

df['Arrival_Time'].str.split(' ')
```

```python
df['Arrival_Time'].str.split(' ').str[0]
```

```python
#using lambda function

df['Arrival_Time']=df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
#Arrival time should be split into hr and min

df['Arrival_hour']=df['Arrival_Time'].apply(lambda x:x.split(':')[0])
df['Arrival_min']=df['Arrival_Time'].apply(lambda x:x.split(':')[1])
```

```python
df.head()
```

```python
df['Arrival_hour']=df['Arrival_hour'].astype('int64')
df['Arrival_min']=df['Arrival_min'].astype('int64')
```

```python
df.info()
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df.head()
```

```python
#Departure time should be split into hr and min

df['Dep_hour']=df['Dep_Time'].apply(lambda x:x.split(':')[0])
df['Dep_min']=df['Dep_Time'].apply(lambda x:x.split(':')[1])
```

```python
df['Dep_hour']=df['Dep_hour'].astype('int64')
df['Dep_min']=df['Dep_min'].astype('int64')
```

```python
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
df.head()
```

```python
#Handling 'Route' Feature
#Route is not a necessarily required in this scenario as Total stops are given.If we try to split the route,lot of null values
#will be there on the columns as the stops are not similar in every case

```

```python
#Handling 'Total_Stops' feature

df['Total_Stops'].unique()
```

```python
#To retrieve the row which has nan value in Total Stops

df[df['Total_Stops'].isnull()]
```

```python
#These are categorical features so we gave to convert into numerical values

# 1st Method
#pd.get_dummies(df['Total_Stops'],drop_first=True)

#2nd Method using map function
df['Total_Stops']=df['Total_Stops'].map({'non-stop':0,'1 stop':1, '2 stops':2,'3 stops':3,'4 stops':4,'nan':1}) #nan is in one sample
df.head()
```

```python
#Dropping the 'Route' feature
df.drop('Route',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Additional_Info'].unique()
```

```python
#Handling 'Duration' Feature

#Departure time should be split into hr and min
#df['Duration'].str.split(' ')
#df['Duration_hr']=df['Duration'].str.split(' ').str[0].str[0] # series splitting


df['Duration_hour']=df['Duration'].str.split(' ').str[0].str.split('h').str[0] # series splitting 

df['Duration_hour'].unique()
```

```python
df[df['Duration']=='5m']
```

```python
#we have to drop the 2 records with duration '5m' as the entries are not valid or wrongly entered the duration 
#or else we cannot convert the Duration_hour into int
df.drop(6474,axis=0,inplace=True)
df.drop(2660,axis=0,inplace=True)
```

```python
df[df['Duration']=='5m']
```

```python
df['Duration_hour']=df['Duration_hour'].astype('int64')
df.info()
```

```python
#splitting into Duration into mins
df['Duration_min']=df['Duration'].str.split(' ').str[1].str.split('m').str[0] # series splitting 
df['Duration_min'].unique()
```

```python
df[df['Duration_min'].isnull()]
```

```python
# Filling the null values with 0 since there is no min value in the duration
df['Duration_min']=df['Duration_min'].fillna(0)
df['Duration_min'].unique()
```

```python
df['Duration_min']=df['Duration_min'].astype('int64')
```

```python
df.info()
```

```python
#Dropping Duration column 
df.drop('Duration',axis=1,inplace=True)
df.head()
```

```python
#Handling other categorical Feature

df['Airline'].unique()
```

```python
# As we are using the data to perform train and test,we use Label Encoder to convert into numerical values

from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
```

```python
df['Airline']=label_encoder.fit_transform(df['Airline'])
df['Source']=label_encoder.fit_transform(df['Source'])
df['Destination']=label_encoder.fit_transform(df['Destination'])
df['Additional_Info']=label_encoder.fit_transform(df['Additional_Info'])
```

```python
#using get_dummies

pd.get_dummies(df,columns=['Airline','Source','Destination','Additional_Info'],drop_first=True)
```
