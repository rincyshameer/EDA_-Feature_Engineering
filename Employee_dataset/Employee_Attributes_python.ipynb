```python
import numpy as np
Dept=['Sales','Research & Development','Human Resources']
Gender=['Male','Female']
result=np.zeros([len(Gender),len(Dept)])
print(result)

for i in range(len(Gender)):
    for j in range(len(Dept)):
        gender=data[(data['Department']==Dept[j]) & (data['Gender']==Gender[i])]
       # display(gender)
        result[i,j]=gender['HourlyRate'].mean()
        #print(result)             
df=pd.DataFrame(result,index=Gender,columns=Dept).T
display(df)
                       
df.plot(kind='bar')                        
```
